```python


import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv').drop(columns=['Survived'])
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
final_df = pd.concat([train,test]).sample(1309)
```

```python
'''
we would be taking a sample from the age
we want to check if the actual mean of age is less than 35 (35 is the claimed mean by someone and we are trying to challenge this hypothesis)
'''
pop = final_df['Age'].dropna()
```

```python
# H0 -> The mean age is 35
# H1 -> The mean is less than 35
```

```python
pop.mean()
'''
this is the true population mean just for reference at last for checking whether the actual mean was 35 (H0 being true) or whether it was less than 35 
(H1 being true)
'''
```

```python
sample_age = pop.sample(25).values #taking 25 people's age, so n = 25 
```

```python
sample_age
```

```python
# check for normality using Shapiro Wilk test
'''
one of the assumptions of t test is that the pop dist should be normal in nature. if we can't know anything about the population, which in real life we 
obviously can't, we check the if the sample data is following normal dist or if its normalish. for that, we use the shapiro wilk test, where in, if the 
p value of the test is greater than the alpha (0.05), the dist follows normal dist. 
'''
from scipy.stats import shapiro

shapiro_age = shapiro(sample_age)

print(shapiro_age)
```

```python
pop_mean = 35
```

```python
import scipy.stats as stats

'''
below, you can directly do the single sample t test, by just feeding sample array and population mean. it will automatically calculate the sample mean, 
sample std deviation required to get the t value and will give you the corresponding p value as well. 
NOTE:- this function below will give you p value always with the assumption that you are doing a 2 tailed test. hence, if you are doing a single tailed
test as above where i am finding if mean age is lesser than claimed mean, i simply divide the p-value by 2. 
'''

t_statistic, p_value = stats.ttest_1samp(sample_age, pop_mean)

print("t-statistic:", t_statistic)
print("p-value:", p_value/2)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
