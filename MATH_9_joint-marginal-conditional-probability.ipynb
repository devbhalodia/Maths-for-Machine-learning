```python
import numpy as np 
import pandas as pd 


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
final_df = pd.concat([train,test]).sample(1309)
```

```python
pd.crosstab(final_df['Survived'],final_df['Pclass'])

'''
creates a contigency table of the features specified.
'''
```

```python
pd.crosstab(final_df['Survived'],final_df['Pclass'], normalize='all')

'''
gives joint probability of each cell. 
'''
```

```python
pd.crosstab(final_df['Survived'],final_df['Pclass'], normalize='all', margins=True)

'''
gives joint prob as well as marginal prob which is inside the All column and row. 
for eg: marg prob of survived = 0 is 0.616
'''
```

```python
pd.crosstab(final_df['Survived'],final_df['Pclass'], normalize='columns')

'''
when normalize is set to columns, we are finding the conditional prob of row feature, given the column feature. 
for eg: 0.75 below, represents prob of passenger dying, given he/she is in pclass = 3. 
'''
```

```python
pd.crosstab(final_df['Survived'],final_df['Pclass'], normalize='index')

'''
normalize = index means conditional prob of the column, given the row. for eg:- 
0.677 represents conditional prob of person belonging to pclass = 3, given he/she died. 
'''
```
